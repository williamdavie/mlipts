First we set up a number of LAMMPS calculations for defining a configuration space.

For this to work you must have the directory ```lammps_base_*``` in your working directory. These settings are copied to all lammps caculations. An example is given for PuO2

Notice in in.PuO2 there is a variable lammps will not regonise: ``` £T ```. The code regonises ```£``` as a variable and will generate a directory for each of these values. That way we can set up a NPT caculations at temperatures 100K, 200K etc. The code will recognise multiple variables and will set up N_var1*N_var2... caculations.

In the example we stick with ```£T 100 200```.


If you are running on an hpc, you may want to generate a bash script to run these calculations sequentially. This is done via the option ```run_script=True```, if set to true you then provide a command string ```lammps_cmd_line="srun lmp"``` the line should end with a lmps executable.


````python
from pytrain.training_workflow import trainingWorkflow

workflow = trainingWorkflow('./lammps_base_PuO2','./vasp_base_PuO2')

lammps_exec = "/work/e89/e89/wd324/Programs/lammps/build/lmp"
lammps_cmd_line = "srun --distribution=block:block --hint=nomultithread " + lammps_exec

workflow.build_lammps_calculations()
````
